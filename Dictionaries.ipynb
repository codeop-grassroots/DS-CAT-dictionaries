# Diccionaris

Les cadenes i les llistes són estructures de dades ordenades, la qual cosa implica que es poden indexar. No totes les dades estan ordenades de manera natural, per això Python ofereix diccionaris com a forma d'estructurar dades no ordenades.

## Conceptes bàsics dels diccionaris

Els diccionaris són col·leccions d'elements *no ordenats*. Per identificar un element, cal conèixer la "clau" de l'element.

Bàsicament, els diccionaris emmagatzemen "parells clau-valor". Qualsevol tipus de dades immutable pot funcionar com a clau. Un tipus molt comú d'utilitzar com a clau és la cadena.

Creeu diccionaris amb claudàtors (`{}`), de manera similar a com creeu llistes amb claudàtors. Un diccionari buit que creeu assignant `{}` a una variable. Podeu crear un diccionari amb continguts descrivint cada element del diccionari entre claudàtors utilitzant la sintaxi `<clau>:<valor>`, i comes entre els elements.

Aquí es crea un diccionari "fruitbasket" que conté tres parells clau-valor, és a dir, la clau "poma" amb el valor 3, la clau "plàtan" amb el valor 5 i la clau "cirera" amb el valor 50.

image.png

```python
fruitbasket = { "apple":[3,4], "banana":[5,5], "cherry":[50,4] }

print(fruitbasket)
```

Per accedir al valor que pertany a una clau específica, utilitzeu la mateixa sintaxi que utilitzaríeu per a una llista, excepte que escriviu la clau al lloc on escriureu l'índex d'una llista.

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50 }
print(fruitbasket)
print( fruitbasket["apple"] )
```

```python
list_basket = list(fruitbasket.values())
list_basket[0]
```

Podeu utilitzar un bucle `for` per recórrer un diccionari. La variable del bucle `for` té accés a totes les claus.

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50 }

def traverse_dictionary(dictionary):
    content = ""
    for key in dictionary:
        content += key + ":" + str(dictionary[key])+ "\n"
    return content

print(traverse_dictionary(fruitbasket))
```

Si intenteu accedir a un element del diccionari amb una clau que no està disponible al diccionari, es produirà un error d'execució. Tanmateix, podeu afegir un nou element a un diccionari simplement assignant un valor a l'element del diccionari identificat amb la nova clau. Per exemple, afegint un "mango" a la cistella de fruites, podeu fer el següent:

```python
fruitbasket = { "apple": 3, "banana": 5, "cherry": 50 }
print( fruitbasket )
fruitbasket["mango"] = 3
```

```python
fruitbasket["mango"] += 3
print( fruitbasket )
```

Sobreescriure un element de diccionari funciona exactament de la mateixa manera que crear un element de diccionari nou: només li assigneu un valor.

Esborrant un element d'un diccionari, feu servir la paraula clau `del`, de la mateixa manera que podeu utilitzar amb les llistes.

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50 }
print( fruitbasket )
del fruitbasket["banana"]
print( fruitbasket )
```

Per tant, els diccionaris són força limitats, però tenen els seus usos.

## Mètodes de diccionari

Aquesta secció descriu els mètodes de diccionari que s'utilitzen més sovint.

### `copy()`

Igual que les llistes, si assigneu una variable que conté un diccionari a una altra variable, no esteu creant una còpia del diccionari; en realitat esteu creant un àlies (si no recordeu què és un àlies, consulteu el capítol anterior). No podeu fer servir el truc que s'utilitza per a les llistes per crear una còpia, ja que utilitza una sintaxi de slice i els diccionaris no admeten slices. Per tant, hi ha un mètode `copy()` que retorna una còpia d'un diccionari.

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50 }
fruitbasketalias = fruitbasket
fruitbasketcopy = fruitbasket.copy()

print( id( fruitbasket ) )
print( id( fruitbasketalias ) )
print( id( fruitbasketcopy ) )
```

Tingueu en compte que aquest mètode fa una còpia superficial del diccionari (vegeu el capítol anterior si no recordeu la diferència entre còpies superficials i profundes). Si voleu fer una còpia profunda, utilitzeu la funció `deepcopy()` del mòdul `copy`.

### `keys()`, `values()`, and `items()`

El mètode `keys()` proporciona un iterador que enumera totes les claus d'un diccionari. El mètode `values()` proporciona un iterador que enumera tots els valors d'un diccionari. El mètode `items()` proporciona un iterador que enumera tots els parells clau-valor d'un diccionari com a tuples.

Dic específicament que aquests mètodes retornen un iterador i no una llista. Si voleu convertir-los en llistes, heu d'utilitzar el càsting de llista.

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50 }
print( list(fruitbasket.keys() ) )
print( list( fruitbasket.values() ) )
print( list( fruitbasket.items() ) )
```

En aquest punt, potser us preguntareu quan podeu utilitzar un iterador. Utilitzeu principalment iteradors per a bucles `for` (tot i que també podeu utilitzar-los com a arguments per a les funcions `max()`, `min()` i `sum()`).

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50, "durian":0, "mango":2 }
fruitbasket["lemon"] = 10
for key in fruitbasket.keys():
    print( "{}:{}".format( key, fruitbasket[key] ) )
    
print( sum( fruitbasket.values() ) )
```

No podeu aplicar el mètode `sort()` al càsting de la llista, és a dir, `keylist = list( fruitbasket.key() ).sort()` no funciona. Primer heu de crear la llista i després ordenar-la. Tampoc podeu escriure `for key in keylist.sort()`, ja que el mètode `sort()` no té valor de retorn.

Si us pregunteu per què sembla que Python prefereix els iteradors en lloc de les llistes: la resposta és que els iteradors són més generals i utilitzen molta menys memòria. Són mètodes "lazy", ja que només proporcionen un element quan se'l demana.

### `get()`

El mètode `get()` es pot utilitzar per obtenir un valor d'un diccionari fins i tot quan no sabeu si la clau per a la qual busqueu el valor existeix. Truqueu al mètode `get()` amb la clau que esteu cercant, i retornarà el valor corresponent quan la clau existeixi al diccionari, o el valor especial `None` quan la clau no existeix al diccionari. Si voleu retornar un valor específic en lloc de `None` si la clau no existeix, podeu afegir aquest valor com a segon argument al mètode.

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50, "durian":0, "mango":2 }
print(fruitbasket.get("love",0))
print(fruitbasket["apple"])
print(fruitbasket)
```

```python
fruitbasket = { "apple":3, "banana":5, "cherry":50, "durian":0, "mango":2 }


def keyvalue_exists(fruitbasket, item): 
    if fruitbasket.get( item ):
        return (item + " is in the basket" )
    else:
        return ("no "+ item + " in the basket")
        
print(keyvalue_exists(fruitbasket, "apple"))
print(keyvalue_exists(fruitbasket, "orange"))

banana = fruitbasket.get( "banana", 0 )
print( "number of bananas in the basket:", banana )

strawberry = fruitbasket.get( "strawberry", 0 )
print( "number of strawberries in the basket:", strawberry )
```

Estudieu de prop l'exemple anterior, ja que el que demostra sobre el mètode `get()` és molt útil. Suposem que emmagatzemeu una col·lecció d'articles amb les quantitats corresponents, per exemple, el contingut d'una cistella de fruites amb les claus els noms de les fruites i els valors les quantitats. Quan consulteu el diccionari de fruitbasket mitjançant el mètode `get()` amb un segon paràmetre zero, podeu buscar qualsevol fruita a la cistella sense necessitat de comprovar primer si la fruita existeix a la cistella, perquè si demaneu una fruita això no hi és, el mètode `get()` retorna zero, que és exactament el que voleu escoltar.

## Storing complicated values

Fins ara només pensava en el cas en què un diccionari emmagatzema un únic valor d'un tipus de dades simple. Tanmateix, és possible emmagatzemar valors molt més complexos als diccionaris. Els valors poden ser objectes Python arbitraris. Per exemple, podeu emmagatzemar una llista amb cada clau. A sota s'utilitza un diccionari per emmagatzemar els estudiants que segueixen un curs. El curs s'identifica pel seu número de curs, mentre que els estudiants s'identifiquen pel seu número d'estudiant.

```python
courses = {
    '880254':['u123456', 'u383213', 'u234178'], 
    '822177':['u123456', 'u223416', 'u234178'], 
    '822164':['u123456', 'u223416', 'u383213', 'u234178']}

for c in courses:
    print( c )
    
    for s in courses[c]:
        print( s, end=" " )
        
    print()
    print(" ")
```

Suposem que no només volem emmagatzemar els números d'estudiant d'un número de curs, sinó també el nom del curs, el valor ECTS del curs i per a cada número d'estudiant també la nota. Podeu fer-ho (per exemple) emmagatzemant el valor d'un número de curs com a diccionari, amb tres claus, és a dir, "nom", "ects" i "estudiants". El valor de "nom" és el nom del curs com a cadena, el valor de "ects" és l'ECTS com a nombre enter i el valor de "estudiants" és un altre diccionari, que conté números d'estudiant com a claus i qualificacions com a valors.

```python
courses = {
    '880254': 
    { "name":"Research Skills Data Processing", "ects":3, 
      "students":{'u123456':8, 'u383213':7.5, 'u234178':6} }, 
    '822177': 
    { "name":"Understanding Intelligence", "ects":6,
     "students":{'u123456':5, 'u223416':7, 'u234178':9} }, 
    '822164': 
    { "name":"Computer Games", "ects":6,
     "students":{'u123456':7.5, 'u223416':9, 'u383213':6, 'u234178':4} } }

for c in courses:
    print( "{}: {} ({})".format( c, courses[c]["name"], courses[c]["ects"] ) )
    for s in courses[c]["students"]:
        print( "{}: {}".format( s, courses[c]["students"][s] ) )
    print()
```

## Lookup speed

Les llistes i els diccionaris són les dues estructures de dades més utilitzades a Python. Tot i que sovint està clar quan hauríeu d'utilitzar quina estructura de dades, és útil saber una mica sobre com processa Python aquestes estructures de dades en cas que tingueu una opció.

Suposem que llegim una gran quantitat de números d'un fitxer. Els números són tots diferents i poden ser qualsevol cosa. Més tard hem de comparar els números d'una altra llista amb els números que llegim del fitxer.

Hem d'utilitzar una llista o un diccionari per emmagatzemar els números que llegim del fitxer? Com que només són números, sense dades addicionals, una llista sembla ser la millor opció. Tanmateix, hi ha un problema si feu servir una llista aquí. Consulteu el codi següent, en el qual es crea una llista de 10.000 números, i després algun codi comprova si hi ha 10.000 números diferents si estan a la llista (que cap d'ells).

```python
from datetime import datetime


def duration_check_numbers_in_list():
    numlist = []

    for i in range( 10000 ):
        numlist.append( i )

    start = datetime.now()

    count = 0
    for i in range( 10000, 20000 ):
        if i in numlist:
            count += 1

    end = datetime.now()
    return( "{}.{} seconds needed to find {} numbers".format( 
            (end - start).seconds, (end - start).microseconds,count ) )
print(duration_check_numbers_in_list())
```

Aquí teniu el codi per fer el mateix amb un diccionari, on simplement emmagatzemem el valor 1 amb cada número.

```python
from datetime import datetime

def duration_check_numbers_in_dictionary():
    numdict = {}

    for i in range( 10000 ):
        numdict[i] = 1

    start = datetime.now()

    count = 0
    for i in range( 10000, 20000 ):
        if i in numdict:
            count += 1

    end = datetime.now()

    return( "{}.{} seconds needed to find {} numbers".format( 
            (end - start).seconds, (end - start).microseconds,count ) )
print(duration_check_numbers_in_dictionary())
```

Notareu que per a un diccionari, el codi dóna una resposta gairebé immediatament, mentre que per a una llista triga força temps perquè el codi doni una resposta.

El motiu és que faig servir l'operador `in` per comprovar si un número és a la llista o al diccionari. Per a una llista, això significa que Python cerca a través de la llista, seqüencialment, fins que arriba al número o arriba al final de la llista. En aquest cas, vol dir que Python comprova 10.000 vegades 10.000 números (ja que no en troba cap), que són 100 milions de números.

Per a un diccionari, el procés de trobar una clau és molt més ràpid. Python pot decidir ràpidament si una clau està o no en un diccionari. Normalment, només n'hi ha prou amb la comprovació d'un grapat de números. Per tant, el codi és molt, molt més ràpid per a un diccionari.

Podríeu pensar que un parell de segons per a la cerca de la llista encara són insignificants, però el temps de cerca augmenta quadràticament amb la mida de les dades. Depenent del problema, pot ser molt preferible utilitzar un diccionari en lloc d'utilitzar una llista.

D'altra banda, les llistes ocupen menys memòria que els diccionaris, i si podeu accedir directament a un element de llista mitjançant el seu índex, les llistes són més ràpides que els diccionaris. Per exemple, al problema anterior, si la llista està ordenada, podeu trobar-hi números d'una manera més intel·ligent que amb l'operador `in` (comprovant uns 14 índexs n'hi hauria prou); en aquest cas, una llista pot tornar a ser més ràpida. .

A partir d'això, hauríeu de recordar que una llista és ràpida si podeu accedir als seus elements directament mitjançant el seu índex, mentre que un diccionari és una opció molt millor si la forma principal de trobar alguna cosa és escanejant elements. L'operador `in` sembla fàcil i llegeix bé, però si l'utilitzeu per buscar alguna cosa en una llista llarga, millor que ho penseu de nou.
